```python
# Count missing values per column
missing_values = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:")
print(missing_values)
```

```python
# Standardize column names: lowercase, replace spaces and special characters
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('.', '').str.replace('/', '_or_')
df.columns
```

```python
df.rename(columns={"team": "club"}, inplace=True)
df.rename(columns={"club_league_name": "league"}, inplace=True)
df.rename(columns={"nationality": "national_team"}, inplace=True)
```

```python
# Replace 'Unknown' with NaN
df.replace('Unknown', pd.NA, inplace=True)
df.tail()
```

```python
# Clean height: Extract cm (e.g., '169cm / 5'7"' -> 169)
def clean_height(height):
    if pd.isna(height):
        return pd.NA
    match = re.search(r'(\d+)cm', str(height))
    return int(match.group(1)) if match else pd.NA

# Clean weight: Extract kg (e.g., '67kg / 148lbs' -> 67)
def clean_weight(weight):
    if pd.isna(weight):
        return pd.NA
    match = re.search(r'(\d+)kg', str(weight))
    return int(match.group(1)) if match else pd.NA

# Apply cleaning and rename columns
df['height_cm'] = df['height'].apply(clean_height)
df['weight_kg'] = df['weight'].apply(clean_weight)
df.drop(columns=['height', 'weight'], inplace=True)  
```

```python

# Convert numeric columns to appropriate type, excluding kit numbers
numeric_cols = [
    'age', 'overall_rating', 'potential', 'height_cm', 'weight_kg', 'total_attacking', 'crossing', 
    'finishing', 'heading_accuracy', 'short_passing', 'volleys', 'total_skill', 'dribbling', 
    'curve', 'fk_accuracy', 'long_passing', 'ball_control', 'total_movement', 'acceleration', 
    'sprint_speed', 'agility', 'reactions', 'balance', 'total_power', 'shot_power', 'jumping', 
    'stamina', 'strength', 'long_shots', 'total_mentality', 'aggression', 'interceptions', 
    'att_position', 'vision', 'penalties', 'composure', 'total_defending', 'marking', 
    'standing_tackle', 'sliding_tackle', 'total_goalkeeping', 'gk_diving', 'gk_handling', 
    'gk_kicking', 'gk_positioning', 'gk_reflexes', 'total_stats', 'base_stats', 'weak_foot', 
    'skill_moves', 'international_reputation'
]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Convert kit numbers to nullable integers
kit_cols = ['club_kit_number', 'national_team_kit_number']
for col in kit_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')
```

```python
df[numeric_cols]
```

```python
# Clean currency columns
def clean_currency(value):
    if pd.isna(value) or value == '0':
        return 0
    value = str(value).replace('€', '').strip()
    multiplier = 1000000 if 'M' in value else 1000 if 'K' in value else 1
    value = value.replace('M', '').replace('K', '')
    return float(value) * multiplier if value.replace('.', '').isdigit() else pd.NA

for col in ['value', 'wage', 'release_clause']:
    df[col] = df[col].apply(clean_currency)
```

```python
# Convert joined to datetime (YYYY-MM-DD)
def clean_joined(value):
    if pd.isna(value):
        return pd.NA
    if isinstance(value, pd.Timestamp):
        return value
    try:
        # Handle Excel serial dates (e.g., 37987 -> 2004-01-01)
        return datetime(1899, 12, 30) + timedelta(days=int(float(value)))
    except (ValueError, TypeError):
        # Handle string dates or invalid values
        return pd.to_datetime(value, errors='coerce')

# Apply cleaning and convert to datetime, then format as YYYY-MM-DD
df['joined'] = pd.to_datetime(df['joined'].apply(clean_joined), errors='coerce').dt.strftime('%Y-%m-%d')
```

```python
# Split contract_dates and convert to integers
df[['contract_start', 'contract_end']] = df['contract_dates'].str.split(' ~ ', expand=True)
df['contract_start'] = pd.to_numeric(df['contract_start'], errors='coerce').astype('Int64')
df['contract_end'] = pd.to_numeric(df['contract_end'], errors='coerce').astype('Int64')
df.drop(columns=['contract_dates'], inplace=True)
```

```python
# Add contract_duration
df['contract_duration'] = (df['contract_end'] - df['contract_start']).astype('Int64')
```

```python
# Clean foot and body_type
df['foot'] = df['foot'].str.capitalize()
df['body_type'] = df['body_type'].str.split(' \\(').str[0].str.strip()
```

```python
# Clean player_specialities: Split into list, then convert to string for deduplication
df['position'] = df['position'].str.split(', ').apply(lambda x: x if isinstance(x, list) else [])
df['position'] = df['position'].apply(lambda x: ', '.join(x) if isinstance(x, list) and x else pd.NA)
df['position']
```

```python
# Clean player_specialities: Split into list, then convert to string for deduplication
df['player_specialities'] = df['player_specialities'].str.split(', ').apply(lambda x: x if isinstance(x, list) else [])
df['player_specialities'] = df['player_specialities'].apply(lambda x: ', '.join(x) if isinstance(x, list) and x else pd.NA)
df['player_specialities']
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# 1. Filter young stars (age ≤ 23, potential ≥ 87)
df_young_stars = df[(df['age'] <= 23) & (df['potential'] >= 87)][['name', 'fifa_version_name', 'age', 'potential', 'value', 'best_position', 'national_team', 'club']]
df_young_stars
```

```python
# 1. Sorting: Sort by name, value descending within each FIFA version
df_sorted = df.sort_values(['fifa_version_name', 'name','value'])
df_sorted
```

```python
# top leagues
top_leagues = ['La Liga', 'Premier League', 'Bundesliga', 'Serie A', 'Ligue 1']

# Grouping and Summarizing: Find highest-rated player by league and version
df_top_leagues = df[df['league'].isin(top_leagues)]
df_highest_rated = df_top_leagues.groupby(['fifa_version_name', 'league']).apply(
    lambda x: x.loc[x['overall_rating'].idxmax(), ['name', 'club', 'overall_rating', 'value', 'wage']]
)
df_highest_rated
```

```python
# Filtering Rows: Subset elite players (overall_rating >= 87) in top leagues
df_elite_players = df[(df['league'].isin(top_leagues)) & (df['overall_rating'] >= 87)][
    ['name', 'fifa_version_name', 'league', 'overall_rating', 'value', 'wage', 'best_position', 'national_team']
]
df_elite_players
```

```python
# 2. Filter and Groupby: Mean value, wage, rating for elite players (overall_rating >= 87)
df_elite_players = df[(df['league'].isin(top_leagues)) & (df['overall_rating'] >= 87)]
df_elite_players_agg = df_elite_players.groupby(['fifa_version_name', 'league']).agg({
    'value': 'mean',
    'wage': 'mean',
    'overall_rating': 'mean'
}).reset_index()
df_elite_players_agg.columns = ['fifa_version_name', 'club_league_name', 'mean_value', 'mean_wage', 'mean_overall_rating']

df_elite_players_agg
```

```python
# 4. Pivot Table: Summarize value and wage by version and league
df_pivot = pd.pivot_table(
    df[df['league'].isin(top_leagues)],
    index=['fifa_version_name', 'league'],
    values=['value', 'wage'],
    aggfunc=[np.mean, np.median, min, max, np.std]
)
df_pivot
```

```python
# Sort: By wage in top leagues
df_sorted_wage = df[df['league'].isin(top_leagues)].sort_values(
    ['fifa_version_name', 'wage'])
df_sorted_wage
```

```python
# Select Subset: Financial columns
financial_cols = ['name', 'fifa_version_name', 'league', 'overall_rating', 'value', 'wage', 'club']
df_financial_subset = df[df['league'].isin(top_leagues)][financial_cols]
df_financial_subset
```
